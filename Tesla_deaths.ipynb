Problem 1: Warm Up – One-Dimension Random Walk along a line in a 2-D plane
Imagine a person walking down a road from point ‘A’ to point ‘B’. Ideally one expects that person would walk in a straight line. However, the person walks such that every step she moves forward, she also moves one step to the right or left with equal probability. Write a Python code to simulate this walk and plot it as follows:

a.	The user should input the number of steps, and the code should plot the position of the walker v.s. the step number.

b.	Run a loop that does the random walk simulation 100 times and for 20 steps for each simulation. Store the position of the walker in a list or an array for each step of each simulation. Plot the average position of the walker for each of the 20 steps taken, along with error bars. As the number of steps taken increases where is the walker most likely to be found at?

c.	Redo part(b) for the case with bias, i.e., where the probability of moving one step to the left slightly more – e.g., 5% more – than the probability of moving one step to the right.


```python
import numpy as np
import matplotlib.pyplot as plt
#The user should input the number of steps, and the code should plot the position of the walker v.s. the step number

n=int(input("enter the no.of steps"))


position = 0
positions = [position]

for i in range(n):
    step = np.random.choice([-1, +1])
    position = position + step
    positions.append(position)
full_pos=0

print (positions)
plt.plot(range(n + 1), positions)
plt.xlabel("step number")
plt.ylabel("position of the walker")
plt.title ("sample plot")
```

```python
#d. From the description column, determine a way to describe the cause of highest accidents using commonly occurring words
import pandas as pd
from collections import Counter

df=pd.read_csv('Tesla Deaths - Deaths.csv')
words = ["hit", "hits", "drives", "collision", "collides", "crashes", "runs", "strikes", "kills","accident","crashed","crash","rear","burns"]
l=[]
word_count=Counter()
for description in df[' Description ']:
  if pd.notna(description):
    found=False
    new_words = description.lower().split()
    for new_word in new_words:
      if new_word in words:
        found=True
        word_count[new_word]+=1
        break
    if found:
      l.append(description)
print(l)
for new_word, count in word_count.items():
    print(f"{new_word}: {count}")



```

```python
#e. Determine the number of cyclist related fatalities by country.Does this give a measure of which country is most unsafe for cyclists?
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('Tesla Deaths - Deaths.csv')

df['Cyclists'] = df[' Cyclists/ Peds '].str.replace("-","0").fillna(0).astype(int)
cyclists_deaths = df.groupby([' Country '])['Cyclists'].sum()
print(cyclists_deaths)
plt.bar(cyclists_deaths.index, cyclists_deaths.values)
plt.xticks(rotation=90)
plt.show()
```

```python
#f. What fraction of Autopilot claimed accidents are actually verified?
#Is there a correlation between autopilot related accidents to any of the other factors based on the provided data?
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('Tesla Deaths - Deaths.csv')
Verified=df[" Verified Tesla Autopilot Deaths "].str.replace("-","0").fillna(0).astype(int).sum()/df[" Autopilot claimed "].str.replace("-","0").fillna(0).astype(int).sum()
print(Verified)#to obtain correlation

column=[" Tesla driver "," Tesla occupant "," Other vehicle "," Cyclists/ Peds "," Deaths "]
df[" Autopilot claimed "]=df[" Autopilot claimed "].str.replace("-","0").fillna(0).astype(int)
Correl=df[column].astype(int).corrwith(df[" Autopilot claimed "])
print(Correl)
```

```python
# g. Determine the ‘model’-wise distribution of accidents. Which model results in minimal damage to the driver?
import pandas as pd
df1=pd.read_csv("Tesla Deaths - Sudden Acceleration.csv")
models = df1['Model'].value_counts()
ax = models.plot.bar( figsize=(7,6))
print(models)
plt.xlabel(' Model ')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents vs Model')
plt.xticks(rotation=90)
plt.show()
print("Minimal damage:", models.index[-3:])
```

```python
# h. Using the ‘Tesla Deaths – Sudden Acceleration.csv’ find the distribution of accidents by model and by State (note, the locations are given as “city, state”. You have to extract state information). Do you find any difference in the distribution of accidents attributed to different ‘models’ for sudden acceleration and accidents?

import matplotlib.pyplot as plt
import pandas as pd
df1=pd.read_csv("Tesla Deaths - Sudden Acceleration.csv")
df1['State'] = df1['Location'].str.split(',')
df1['State'] = df1['State'].str[1]#.str.strip()
print (df1['State'])
modelvsstate = pd.crosstab(df1['Model'], df1['State']).fillna(0)

ax = modelvsstate.plot(kind='bar', stacked=True, figsize=(15, 4))
plt.xlabel('Model')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Model vs State')
plt.xticks(rotation=90)
plt.legend(title='State')
plt.show()
```
